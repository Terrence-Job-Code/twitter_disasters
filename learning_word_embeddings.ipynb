```python
import nltk
from nltk.corpus import brown
from gensim.models import Word2Vec
from gensim.models.word2vec import Text8Corpus
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, LabelSet
```

```python
#nltk.download('brown')
print(brown.sents())
```

```python
w2v_model = Word2Vec(brown.sents(),size=128,window=5,min_count=3,workers=4)
```

```python
print(w2v_model.wv['Italy'],w2v_model.wv['France'])
```

```python
print(w2v_model.wv.most_similar('Paris'))
```

```python
print(w2v_model.wv.most_similar(positive=['woman','king'],negative=['man']))
print()
print(w2v_model.wv.most_similar(positive=['Rome','France'],negative=['Italy']))
```

```python
# trying a different corpus (larger)
w2v_model2 = Word2Vec(Text8Corpus('data_for_nlp_toy/text8/text8'), size=100, window=5, min_count=150, workers=4)
```

```python
print(w2v_model2.wv.most_similar('paris'))
```

```python
print(w2v_model2.wv.most_similar(positive=['woman','king'], negative=['man'], topn=3))
print()
print(w2v_model2.wv.most_similar(positive=['girl','father'], negative=['boy'], topn=3),'\n')
print(w2v_model2.wv.most_similar(positive=['paris','italy'], negative=['france'], topn=3),'\n')

print(w2v_model2.wv['woman'])
```

```python
# this just looks at a plot of the word vectors, shows some kind of relationship between the words
# and is used for the word vectors
output_notebook()

X = []
for word in w2v_model2.wv.vocab:
    X.append(w2v_model2.wv[word])
    
X = np.array(X)
print('Computerd X:', X.shape)
X_embedded = TSNE(n_components=2, n_iter=250, verbose=2).fit_transform(X)
print('Computed t-SNE', X_embedded.shape)

df = pd.DataFrame(columns=['x', 'y', 'word'])
df['x'], df['y'], df['word'] = X_embedded[:,0], X_embedded[:,1], w2v_model2.wv.vocab

source = ColumnDataSource(ColumnDataSource.from_df(df))
labels = LabelSet(x='x', y='y', text='word', y_offset=8,
                 text_font_size='8pt', text_color='#555555',
                 source=source, text_align='center')

plot = figure(plot_width=600, plot_height=600)
plot.circle('x', 'y', size=12, source=source, line_color='black', fill_alpha=0.8)
plot.add_layout(labels)
show(plot, notebook_handle=True)
```
